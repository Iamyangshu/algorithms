```python
%matplotlib inline
%config InlineBackend.figure_fromat='retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname=r'C:\Windows\Fonts\simkai.ttf',size=14)
t=np.arange(1,10,0.05)
x=np.sin(t)
y=np.cos(t)
plt.figure(figsize=(5,5))
plt.plot(x,y,'b-')
plt.axis('equal')
#ax1.tick_params(axis='x',width=0.5,colors='glod')
plt.xlabel('正弦',fontproperties=font)
plt.ylabel('余弦',fontproperties=font)
plt.title('一个圆',fontproperties=font)
plt.show()
```
