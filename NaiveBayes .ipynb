```python
import pandas as pd
import numpy as np
from collections import Counter,defaultdict
#读取西瓜数据集
df=pd.read_table(r'E:\Repository\algorithms\xiguabayesclassify.txt',sep=',')
#取出属性
feature=df.iloc[:,:-1]
#取出类别
label=df.iloc[:,-1]
#按类别进行分组，返回的是元祖，feature_grouped[0]是类别，[1]是dataframe有类别所分的组
feature_grouped=list(feature.groupby(by=label))
label_count=Counter(label)
#先验概率和条件概率
prior_probability={}
all_cond_probability=defaultdict(int)


def prior(label):
    for label_name in np.unique(label):
        prior_probability[label_name]=label_count[label_name]/len(label)
    return prior_probability
    
     
#prior(label)
#prior_probability


def div_cond_probability(feature,label):
    #离散属性值的条件概率
    #这里的feature代表某一类下面的某一属性集合
    #计算每一个属性下面特征值对应的个数
    label_value=Counter(feature)
    for each in np.unique(feature):
        all_cond_probability[each+'|'+label]=label_value[each]/len(feature)
    return all_cond_probability

def con_cond_probability(value,feature,label):
    #连续属性值的条件概率
    #这里的feature代表某一类下面的某一属性集合
    #计算每一个属性下面特征值对应的个数
    u=feature.mean()
    eta=feature.std()
    all_cond_probability[feature.name+'|'+label]=(1/(np.sqrt(2*np.pi)*eta))*np.exp(-((value-u)**2/(2*eta**2)))
    return all_cond_probability
        
#cond_probability(feature_grouped[0][1].iloc[:,0])
#cond_probability(feature_grouped[0][1].iloc[:,1])




if __name__=='__main__':

    for each in feature_grouped:
        for i in range(6):
            div_cond_probability(each[1].iloc[:,i],each[0])
        con_cond_probability(0.460,each[1].iloc[:,7],each[0])
        con_cond_probability(0.697,each[1].iloc[:,6],each[0])

    prior(label)        
    #feature_grouped[0][1].shape[1]   
#测试
    test=['青绿','蜷缩','浊响','清晰','凹陷','硬滑','密度','含糖率']
    probability_Good=1
    probability_Bad=1
    for each in test:
        probability_Good=probability_Good*all_cond_probability[each+'|'+'是']
        probability_Bad=probability_Bad*all_cond_probability[each+'|'+'否']
    probability_Good=probability_Good*prior_probability['是']
    probability_Bad=probability_Bad*prior_probability['否']
    if probability_Good>probability_Bad:
        print('判定该样本是好瓜')
    else:
        print('判定该样本为坏瓜')

    


        
    
```
